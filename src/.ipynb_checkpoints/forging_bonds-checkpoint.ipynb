```python
import os
import time

def tap(x,y):
    os.system("adb shell input tap "+str(x)+" "+str(y))
    
def stamina():
    print("Stamina")
    tap(500, 100)
    tap(300, 1000)
```

```python
def forging_bonds(n):
    for i in range(1,n+1):
        print("Start", i)
        tap(400, 1000) #difficult: advance
        tap(400, 1400) #fight!
        
#         if i%10==0:
#             stamina()
        
        time.sleep(3)
        tap(800, 1900) #fight!
        time.sleep(3)
        tap(850, 1750) #AB
        tap(400, 1000) #AB
        time.sleep(20)

        print("After Battle")
        tap(1, 1) #stage clear
        time.sleep(3)
        tap(300, 1200) #close points / items
        time.sleep(3)
        
        for i in range(3):
            tap(800, 50) #skip
            tap(300, 1200) #close points / items
        tap(800, 50) #skip
        tap(300, 1500) #close quest
        
        #game over?
        tap(1, 1)
        tap(400, 1100)
        
def tap_battle(n):
    delta = 30
    ini = time.time()
    while time.time() - ini <= 30:
        tap(850, 1750)
```
